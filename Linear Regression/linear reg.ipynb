```python
import pandas as pd
```

```python
data = pd.read_csv("Advertising.csv")
```

```python
data
```

## Dividing Dataset into training and testing

```python
Y = data['Sales']
X = data[data.columns[:3]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size = 0.8, random_state = 0)
```

```python
import numpy as np
```

```python
x_train_mat, y_train_vec = x_train.to_numpy(), y_train.to_numpy()
```

```python
x_train_mat = np.append(np.ones((x_train_mat.shape[0],1)), x_train_mat, axis = 1)
```

```python
x_test_mat = np.append(np.ones((x_test.shape[0],1)), x_test, axis=1)
```

## Closed Form solution for Linear Regression model parameters

```python
XTX_inv = np.linalg.inv(np.matmul(x_train_mat.T, x_train_mat))
XTY = np.matmul(x_train_mat.T, y_train_vec)
Beta = np.matmul(XTX_inv, XTY)
```

## Sklearn Model for Linear regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print("Sklearn model Intercept: ", model.intercept_)
print("Sklearn model coefficients: ", model.coef_)
```

```python
print("Close Form Intercept: ", Beta[0])
print("Close form coefficients: ", Beta[1:])
```

```python
y_pred = model.predict(x_test)
y_closed = np.matmul(x_test_mat, Beta)
```

## Comparing both models: from sklearn and closed form solution

```python
from sklearn.metrics import mean_squared_error
m1 = mean_squared_error(y_test, y_pred)
m2 = mean_squared_error(y_test, y_closed)
```
